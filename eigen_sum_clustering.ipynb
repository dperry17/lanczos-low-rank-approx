# Sum of kronecker products eigen value clustering

## Motivations
The point of this notebook is to track the eigenvalue clustering of the sum of kronecker products (AKA the generalized sylvester equation). The krylov subspace methods are heavily dependent on the eigenvalues/vector of a matrix. There are low rank solutions for the lyapunov and sylvester equations (sum of two kronecker products), but when the number of products $r \geq 3$, the eigen value clustering will show the solvability of the system and the convergence.
